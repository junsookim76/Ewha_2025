```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

%matplotlib inline    
```

```python
data = np.loadtxt("data_exponential.dat")

print(data.shape)
```

```python
tdata = data[:,0]
data1 = data[:,1]
data2 = data[:,2]

print(tdata.shape)
#print(tdata)
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(tdata, data1, label='data1', color='black', linewidth=0,\
         marker='o', markersize=3, markeredgewidth=1, markeredgecolor='black', mfc='crimson')
ax.plot(tdata, data2, label='data2', color='black', linewidth=0,\
         marker='s', markersize=3, markeredgewidth=1, markeredgecolor='black', mfc='royalblue')


```

```python
def func(x, a):
    return np.exp(-a * x)
```

```python

popt, pcov = curve_fit(func,tdata,data2)

print ("Fit: y = exp(- %.3f * x) " %(popt[0]))

fig, ax = plt.subplots(figsize=(8,5))

ax.plot(tdata, data2, label='data1', color='black', linewidth=0,\
         marker='o', markersize=3, markeredgewidth=1, markeredgecolor='black', mfc='crimson')

plt.plot(tdata, func(tdata,*popt), linewidth=2, color="black", linestyle="dashed", label="curve fit")

plt.show()
```
