## 1. Pandas 소개
- Pandas는 Python에서 데이터 처리를 쉽게 해주는 라이브러리입니다.
- 표 형식의 데이터를 다루기에 매우 유용합니다.

````python
import pandas as pd
import matplotlib.pyplot as plt
````

## 2. Series와 DataFrame
- **Series**: 1차원 데이터
- **DataFrame**: 2차원 표 형식 데이터

````python
s = pd.Series([10, 20, 30, 40])
print(s)
print(type(s))
````

````python
data = {
    '이름': ['영희', '철수', '민수'],
    '국어': [90, 85, 80],
    '수학': [88, 90, 95]
}
df = pd.DataFrame(data)
df
````

## 3. CSV 파일 불러오기
`pd.read_csv()`를 사용해서 CSV 파일을 불러올 수 있습니다.

````python
# 예제용 CSV 파일 다운로드 (인터넷 연결 필요)
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
df = pd.read_csv(url)
df.head()
````

## 4. DataFrame 기초 탐색
- `head()`, `tail()`, `info()`, `describe()` 등을 사용

````python
df.tail()
````

````python
df.info()
````

````python
df.describe()
````

## 5. 인덱싱과 슬라이싱
- 열 선택: `df['열이름']`
- 행 선택: `df.loc[]`, `df.iloc[]`

````python
df['species'].unique()
````

````python
df.loc[0]
````

````python
df.iloc[0:5]
````

## 6. 조건으로 필터링
- 특정 조건을 만족하는 행만 선택

````python
df[df['species'] == 'setosa']
````

## 7. 간단한 시각화
- `matplotlib`을 이용한 기본 시각화

````python
df['sepal_length'].hist()
plt.title('Sepal Length Histogram')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
````
